**Python Packages for Data Science**

Python libraries

Son colecciones de funciones y metodos que te permiten hacer acciones sin escribir codigo.

Las liberias usualmente contienen modulos built-in que proveen diferentes funcionalidades que se pueden usar de manera directa.

El curso divide las librerias de data analysis para Python en tres grupos.

**1. Scientific Computing Libraries**

*   **Pandas:** (Data structures & tools)
 * Ofrece estructura de datos y herramientas para manipulacion de datos de manera efectiva, asi como acceso rapido a datos estructurados. El primer instrumento dde Pandas es una tabla de dos dimensiones que consiste en columnas y filas que es llamado **DataFrame** esta diseñado para proveer la funcionalidad de indexar.

*   **Numpy:** (Arrays & matrices)
 * Usa arrays para sus inputs y outputs. Se puedde extender a objetos por medio de matrices y con cambios minimos de codigo, los desarrolladores pueden realizar procesos rapidos de arrays.

*   **SciPy:** (Integrals, solving differential equations, optimization)
 * Incluye funciones para problemas matematicos avanzados 
 
**2. Visualization Libraries in Python**

La visualizacion de datos es una de los mejores metodos para comunicarnos con otros y mostrarles resultados significativos del analisis.


Estas librerias te permiten crear graficas, mapas, etc.

*   **Matplotlib:** (plots & graphs)
 * Es la mas conocida libreria para visualizacion de datos, los graficos son altamente personalizables

*   **Seaborn:** (plots: heat maps, time series, violin plots)
 * Esta basado en Matplotlib y es muy facil de generar graficos.

**3.  Algorithmic Libraries in Python**

*   **Scikit-learn:** (Machine Learning: Regression, clasification,...)
 * Contiene herramientas para modelado estadistico incluyendo regresion, clasificacion, clustering, etc. Esta libreria ha sido construida con NumPy, SciPy and Matplotlib

* **Statsmodels:** (Explore data, estimate statistical models and perform statistical test)
 * Es tambien un modulo en el que Python permite a los usuarios explorar datos y estimar modelos estadisticos y realizar pruebas estadisticas.

**Importing and Exporting Data in Python**

Propiedades:

* Formato: .csv, .json, .xlsx, .hdf
* File path of dataset:

```python
#Importing a CSV into Python

import pandas as pd

#Define file path

url= "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

df = pd.read_csv(url, header = None)
```

```python
#Printing DataFrame in Python

#df prints the entire dataframe (no recomendado para datasets largos)

#df.head(n) Muestra las primeras **n** filas del data frame

df.head(5)
```

```python
#df.tail(n) muestra las ultimas **n** filas del data frame

df.tail(5)
```

```python
#Exporting a Pandas dataframe to CSV
#Preserve progress anytime by saving modified dataset using

path= "C:\Windows\...\automobile.csv"

df.to_csv(path)
```
